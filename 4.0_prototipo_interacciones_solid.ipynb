```python
import requests
from rdflib import Graph

# 1. URL al archivo
POD_URL = "https://mti-2026-ws.solidcommunity.net/private/datos.ttl"

print(f"Haciendo fetch al POD de Solid en: {POD_URL}...")

response = requests.get(POD_URL)

if response.status_code == 200:
    print("¡Archivo extraído con éxito del POD!\n")

    # 2. Instanciar el grafo en memoria
    g = Graph()
    g.parse(data=response.text, format="turtle")

    print(f"Grafo RDF cargado exitosamente. Total de tripletas: {len(g)}")

    # 3. Consulta de prueba: Listar las entidades y a cuántos conciertos han asistido
    query = """
    PREFIX : <http://example.org/>
    PREFIX p: <https://www.wikidata.org/wiki/Property:>

    SELECT ?persona (COUNT(?concierto) AS ?total_conciertos)
    WHERE {
        ?persona p:P1344 ?concierto .
    }
    GROUP BY ?persona
    """

    print("\n--- Resumen de Asistencia a Conciertos ---")
    results = g.query(query)
    for row in results:
        # Limpiamos la URI para mostrar solo el nombre
        nombre = row.persona.split('/')[-1]
        print(f"👤 {nombre}: {row.total_conciertos} conciertos")

else:
    print(f"Error al extraer los datos: {response.status_code}")
```
